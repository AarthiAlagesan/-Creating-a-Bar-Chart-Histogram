```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (replace with correct file path or URL)
df = pd.read_csv('/content/indian population new.csv')  # Use quotes around the file name

# Quick check of the data
print(df.head())

# Plot (Example: assuming 'Population' column)
plt.figure(figsize=(10,6))
sns.histplot(df['Population'], bins=20, kde=True)
plt.title('Distribution of Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
